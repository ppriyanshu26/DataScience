### 1. Right-Angled Triangle

***

*****
```

### 2. Inverted Right-Angled Triangle
```
*****
****
***
**
*
```

### 3. Pyramid (Center-Aligned)
```
    *
   ***
  *****
 *******

### 4. Inverted Pyramid

 *******

### 5. Diamond Pattern
```

```

---

````python
# ‚≠ê STAR PATTERNS (Logic-based)

n = int(input("Enter rows for star patterns: "))

# 1. Right-Angled Triangle
print("\nRight-Angled Triangle:")
for i in range(1, n + 1):
    x = ""
    for j in range(i):
        x += "*"
    print(x)

# 2. Inverted Right-Angled Triangle
print("\nInverted Right-Angled Triangle:")
for i in range(n, 0, -1):
    x = ""
    for j in range(i):
        x += "*"
    print(x)

# 3. Pyramid (Center-Aligned)
print("\nPyramid:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        x += "*"
    print(x)

# 4. Inverted Pyramid
print("\nInverted Pyramid:")
for i in range(n, 0, -1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        x += "*"
    print(x)

# 5. Diamond Pattern
print("\nDiamond Pattern:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        x += "*"
    print(x)
for i in range(n - 1, 0, -1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        x += "*"
    print(x)

# 6. Hollow Pyramid
print("\nHollow Pyramid:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        if j == 0 or j == (2 * i - 2) or i == n:
            x += "*"
        else:
            x += " "
    print(x)

# 7. Hollow Diamond
print("\nHollow Diamond:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        if j == 0 or j == 2 * i - 2:
            x += "*"
        else:
            x += " "
    print(x)
for i in range(n - 1, 0, -1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(2 * i - 1):
        if j == 0 or j == 2 * i - 2:
            x += "*"
        else:
            x += " "
    print(x)
````

````python
# üî¢ NUMBER PATTERNS (Logic-based)

n = int(input("Enter rows for number patterns: "))

# 8. Right-Angled Number Triangle
print("\nRight-Angled Number Triangle:")
for i in range(1, n + 1):
    x = ""
    for j in range(1, i + 1):
        x += str(j)
    print(x)

# 9. Inverted Number Triangle
print("\nInverted Number Triangle:")
for i in range(n, 0, -1):
    x = ""
    for j in range(1, i + 1):
        x += str(j)
    print(x)

# 10. Number Pyramid
print("\nNumber Pyramid:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(1, i + 1):
        x += str(j)
    for j in range(i - 1, 0, -1):
        x += str(j)
    print(x)

# 11. Repeating Numbers in Triangle
print("\nRepeating Numbers in Triangle:")
for i in range(1, n + 1):
    x = ""
    for j in range(i):
        x += str(i)
    print(x)

# 12. Floyd‚Äôs Triangle
print("\nFloyd's Triangle:")
num = 1
for i in range(1, n + 1):
    x = ""
    for j in range(i):
        x += str(num) + " "
        num += 1
    print(x)

# 13. Pascal‚Äôs Triangle
import math
print("\nPascal's Triangle:")
for i in range(n):
    x = ""
    for j in range(i + 1):
        x += str(math.comb(i, j)) + " "
    print(x)

# 14. Palindromic Number Pyramid
print("\nPalindromic Number Pyramid:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(1, i + 1):
        x += str(j)
    for j in range(i - 1, 0, -1):
        x += str(j)
    print(x)

# 15. Number Diamond
print("\nNumber Diamond:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(1, i + 1):
        x += str(j)
    for j in range(i - 1, 0, -1):
        x += str(j)
    print(x)
for i in range(n - 1, 0, -1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(1, i + 1):
        x += str(j)
    for j in range(i - 1, 0, -1):
        x += str(j)
    print(x)
````

````python
# üî† Character Patterns

n = int(input("Enter rows for character patterns: "))

# 16. Alphabet Triangle
print("\nAlphabet Triangle:")
char = 'A'
for i in range(1, n + 1):
    x = ""
    for j in range(i):
        x += chr(ord('A') + j)
    print(x)

# 17. Inverted Alphabet Triangle
print("\nInverted Alphabet Triangle:")
for i in range(n, 0, -1):
    x = ""
    for j in range(i):
        x += chr(ord('A') + j)
    print(x)

# 18. Pyramid of Characters
print("\nPyramid of Characters:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(i):
        x += chr(ord('A') + j)
    for j in range(i - 2, -1, -1):
        x += chr(ord('A') + j)
    print(x)

# 19. Diamond with Characters
print("\nDiamond with Characters:")
for i in range(1, n + 1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(i):
        x += chr(ord('A') + j)
    for j in range(i - 2, -1, -1):
        x += chr(ord('A') + j)
    print(x)
for i in range(n - 1, 0, -1):
    x = ""
    for j in range(n - i):
        x += " "
    for j in range(i):
        x += chr(ord('A') + j)
    for j in range(i - 2, -1, -1):
        x += chr(ord('A') + j)
    print(x)
````

### 24. Spiral Matrix (5x5)
```
 1  2  3  4  5
16 17 18 19  6
15 24 25 20  7
14 23 22 21  8
13 12 11 10  9
```

---

````python
# üåÄ Miscellaneous Patterns

n = int(input("Enter rows for miscellaneous patterns: "))

# 20. Zig-Zag Pattern
print("\nZig-Zag Pattern:")
for i in range(n):
    for j in range(n):
        if (i + j) % 2 == 0:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

# 21. Butterfly Pattern
print("\nButterfly Pattern:")
for i in range(n):
    for j in range(i + 1):
        print("*", end="")
    for j in range(2 * (n - i - 1)):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end="")
    print()

for i in range(n - 1, 0, -1):
    for j in range(i):
        print("*", end="")
    for j in range(2 * (n - i)):
        print(" ", end="")
    for j in range(i):
        print("*", end="")
    print()

# 22. Hourglass Pattern
print("\nHourglass Pattern:")
for i in range(n):
    print(" " * i + "*" * (2 * (n - i) - 1))

for i in range(n - 2, -1, -1):
    print(" " * i + "*" * (2 * (n - i) - 1))

# 23. Snake Number Pattern
print("\nSnake Number Pattern:")
num = 1
for i in range(n):
    for j in range(n):
        print(num, end=" ")
        num += 1
    num += n
    print()

# 24. Spiral Matrix (5x5)
print("\nSpiral Matrix:")
spiral = [[0 for _ in range(n)] for _ in range(n)]
top, bottom, left, right = 0, n - 1, 0, n - 1
num = 1

while top <= bottom and left <= right:
    for i in range(left, right + 1):
        spiral[top][i] = num
        num += 1
    top += 1
    for i in range(top, bottom + 1):
        spiral[i][right] = num
        num += 1
    right -= 1
    if top <= bottom:
        for i in range(right, left - 1, -1):
            spiral[bottom][i] = num
            num += 1
        bottom -= 1
    if left <= right:
        for i in range(bottom, top - 1, -1):
            spiral[i][left] = num
            num += 1
        left += 1

for row in spiral:
    print(" ".join(str(x) for x in row))
````
