# ⚙️ **What is Data Preprocessing?**
* Data Preprocessing transforms raw data ➡️ usable format for analysis. It's key to ensure:
	- ✅ Accurate
	- 💯 Complete
	- ⚖️ Consistent data
* Steps:
	- 🧼 Cleaning
	- 🔄 Transforming
	- ➕ Integrating

---

## 🧰 Tools:
1. 🐍 **Python**:	Easy syntax, rich libraries
2. 📊 **R**:		Statistical computing, graphics

---

## ⚠️ The Reality of Data:
* Incomplete: Missing values
* Noisy: Errors/outliers
* Inconsistent: Conflicting data
* ⚠️ No quality data ➡️ no quality results!

---

## ⚙️ Major Tasks:
1. 🧼 Data Cleaning:
	- Fill missing ➡️, smooth 📉➡️📈, fix outliers 🗑️, resolve 🛠️
2. ➕ Data Integration:
	- Combine databases 🗄️, cubes 🧊, files 📂
3. ⬇️ Data Reduction:
	- Reduce volume, keep insights 👍
4. 🔄 Data Transformation:
	- Smooth, aggregate, generalize, normalize, build attributes

---

## 🔎 Exploratory Data Analysis (EDA)
* EDA helps visualize and explore data for trends 📊.
* Why?
	- 🐞 Detect errors
	- ✅❓ Test assumptions
	- 🤖➡️ Pick models
	- ↔️ See relationships
	- 📐➡️🎯 Estimate effects

---

## 🛠️ Techniques:
* Raw data plots 📊📈
* Box plots 📦, mean/std plots

---

## ➡️ Handling Missing Values:
* Techniques:
	- 🗑️ Remove nulls
	- 🔢➡️ Impute (mean/median/mode)
* Pandas:
	- .isnull(), .dropna(), .fillna(), .replace(), .interpolate()

---

## 🤔 Data Integration Challenges:
* 👤❓ Entity mismatch
* 👯 Redundancy
* 👯👯 Duplicates
* 💥 Conflicting values
* Same field, different names 🏷️ or derived attributes 💰
